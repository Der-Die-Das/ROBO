```python
import pandas as pd
import matplotlib.pyplot as plt
# Reload the new CSV file
file_path = "./errors.csv"
df = pd.read_csv(file_path)

```

```python
# Constants
BLUE_THRESHOLD = 0.7
RED_THRESHOLD = 0.1

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['error'], marker='o', linestyle='-')

# Add threshold lines
plt.axhline(BLUE_THRESHOLD, color='blue', linestyle='--', label=f'Upper Threshold ({BLUE_THRESHOLD})')
plt.axhline(RED_THRESHOLD, color='red', linestyle='--', label=f'Lower Threshold ({RED_THRESHOLD})')

# Labels and title
plt.xlabel('Timestamp (s)')
plt.ylabel('Error')
plt.title('Error Over Time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
